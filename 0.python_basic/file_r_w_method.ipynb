逐行读取时，有两种方式：
1. 使用readline()方法

输出时，`print(line, end='')` 和 `print(line.strip())` 实现的效果相似但不完全相同：
- `print(line, end='')`
    - 行中的所有原始字符（包括行首和行尾的空格/制表符）
    - 移除了行尾默认的换行符（因为`end=''`覆盖了默认的`end='\n'`）
    - 行尾有换行符（`\n`），它会被保留（因为`readline()`保留了行尾的换行符）
- `print(line.strip())`
    - `strip()`会移除字符串首尾的所有空白字符（包括空格、制表符\t、换行符\n等）
    - 然后`print()`会默认添加一个换行符（因为`end`参数是默认的`\n`）

**关键区别**：
- 如果一行是 `" hello \n"`：
    - `print(line, end='')` → 输出 `" hello "`（保留首尾空格）
    - `print(line.strip())` → 输出 `"hello"`（移除首尾空格）并自动换行
- 对于没有首尾空格的行，两者输出看起来可能相同，但：
    - `print(line, end='')` 会保留原始换行位置
    - `print(line.strip())` 会重新换行（因为`strip()`移除了原换行符，但`print()`又加了一个）

```python
# readline() reads one line at a time
with open('data/text.txt', 'r', encoding='utf-8') as f:
    line = f.readline()
    while line:
        print(line, end='')     # keep the source characters at the beginning and end of the line; " hello \n" will be " hello "
        print(line.strip())     # line.strip() will remove the `\n`, `\t` and `spaces` characters at the beginning and end of the line; " hello \n" will be "hello"
        line = f.readline()
    print("\n")
f.close()

with open('data/datafile.csv', 'r', encoding='utf-8') as f:
    line = f.readline()
    while line:
        # print(line, end='')
        print(line.strip())
        line = f.readline()
f.close()
```

```python
# read the file line by line with iteration object
with open('data/text.txt', 'r', encoding='utf-8') as f:
    for line in f:
        print(line, end='')     # keep the source characters at the beginning and end of the line; " hello \n" will be " hello "
        print(line.strip())
    print("\n")
f.close()

with open('data/datafile.csv', 'r', encoding='utf-8') as f:
    for line in f:
        print(line.strip())
f.close()
```

```python
with open('data/test_data.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()[1:]                #skip the first line

    for line in lines:
        line = line.strip()                  # remove the `\n`, `\t` and `spaces` characters at the beginning and end of the line
        if line=="--------------":
            continue
        else:
            print(line)
```

```python
with open('data/datafile.csv', 'r', encoding='utf-8') as f:
    next(f)  # skip the first line
    for line in f:
        print(line.strip())
```

```python
with open('data/datafile.csv', 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        if i == 0:
            continue

        print(line.strip())
        print(line.strip().split(','))  # split the line by comma and print as a list
```
